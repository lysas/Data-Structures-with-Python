```python
def swap(x,y):
    temp = x
    x = y
    y = temp
    return x,y
```

```python
def quickSort(A, low, high):
    print(f"\nQS({low},{high})")
    
    if(low < high): # Till there is only 1 element
        
        #Initialize
        pivot = A[low]
        i = low + 1
        j = high
        
        print(f"i={i},j={j}")
        
        
        
        # Move smaller elements to left side and larger elements to right side
        while(True):
            while(pivot > A[i] and i<high):
                i = i + 1
                print(f"i = {i}")
            while(pivot < A[j] and j>low):             
                j = j - 1
                print(f"j={j}")
            if i<j:
                A[i],A[j] = swap(A[i],A[j])
                print(A)
            else:
                break
        
        # Completion of an iteration
        # pivot goes to right position
        print(A[low],A[j])
        A[low],A[j] = swap(A[low],A[j])
        print(A)
        
        # Make Recursive Call
        # Call on the elements to the left of pivot
        quickSort(A, low, j-1)
        
        # Call on the elements to the right of pivot
        quickSort(A, j+1, high)
```

```python
# Test for even number of elements 
print("*****Even Number of Elements*****")
A_even = [4,1,2,5,6,3,8,7]
print(f"Length = {len(A_even)}")
quickSort(A_even, 0,len(A_even)-1)
print("Sorted List")
print(A_even)
```

```python
# Test for odd number of elements
print("\n*****Odd Number of Elements*****")
A_odd= [4,1,2,5,6,3,8,7,-3]
print(f"Length = {len(A_odd)}")
quickSort(A_odd, 0,len(A_odd)-1)
print("Sorted List")
print(A_odd)
```
